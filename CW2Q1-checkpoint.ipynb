```python
import numpy as np
import matplotlib.pyplot as plt
```

i) The following codes produce the discrete points in $2D$ ($x_k$ versus $t_m$) for with a mesh size of $h = 0.1$ and $\delta t = 0.1h^2$ with the initial conditions in blue colour, the boundary conditions in green and the approximation points in red

```python
P = np.ones([9,9])
p = np.ones(9)
q = np.zeros(9)
r = np.zeros(11)
s = np.ones(11)
x = np.linspace(0.1,0.9,9)
y = np.linspace(0,1,11)
for i in range(0,9):
    P[i,:]=x[i]*P[i,:]
fig = plt.figure()    
for j in range(0,9):
    plt.scatter(P[:,j],0.01*P[j,:], c='red')
plt.scatter(x,0.01*p,c='red')
plt.scatter(x,0.01*q,c='blue')
plt.scatter(r,0.01*y,c='green')
plt.scatter(s,0.01*y,c='green')
plt.title('Discrete point in 2D ($x$ versus $t$)')
plt.xlabel('$x$')
plt.ylabel('$t$')
plt.xlim(-0.1,1.1,11)
plt.ylim(-0.001,0.011,11)
plt.grid()
```

Creating the matrix $A$ of FTCS and find the approximation by iteration

```python
#each row of A is the coefficient of iteration to get u_(n+1) from u_(n,j+1), u_(n,j), u_(n,j-1)

A = np.zeros([9,9])
b = np.zeros(9)
for i in range(0,8):
    A[i,i+1]=0.1
    A[i+1,i]=0.1
for j in range(0,9):
    A[j,j]=0.8
    
#U[i,j] is the approximation value at x_i, t_j

u = np.zeros(9)
U = np.zeros([9,10])
h = 0.1
for k in range(0,9):
    if k < 3:
        u[k] = 3*(k+1)*h
    else:
        u[k] = 3*(1-(k+1)*h)/2
U[:,0] = np.dot(A,u) + b 
for m in range(1,10):
    U[:,m] = np.dot(A,U[:,m-1]) + b
```

ii) The following codes plot the numerical approximation as a function of $x_k$ with each color representing the different time steps $t_m = 0.01mh^2$, $m=1,2,...,10$. It also plots the numerical approximation as colour plot as a function of $x_k$ on the $x$-axis and time on the $y$-axis.

```python
fig = plt.figure() 
for n in range(0,10):
    plt.plot(P[:,0],U[:,n])
plt.grid()
plt.title('the numerical approximation a function of x with each color representing the different time steps')
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend(labels=['$t_1$','$t_2$','$t_3$','$t_4$','$t_5$','$t_6$','$t_7$','$t_8$','$t_9$','$t_{10}$'])
```
